```
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
```

# Create dataset

```
# XOR Dataset

X = np.array([[-1., -1.],
            [-1., 1.],
            [1., -1.],
            [1., 1.]])
y = np.array([[-1.],
             [1.],
             [1.],
             [-1.]])

print(X.shape, y.shape)
```

# Create Model

## Define Model Architecture

```
HIDDEN = 10

model = keras.Sequential([
    keras.layers.Dense(HIDDEN, activation=tf.nn.relu, input_dim=2),
    keras.layers.Dense(1)
])
```

## Train Model

```
model.compile(optimizer=keras.optimizers.SGD(lr=0.1),
              loss='mean_squared_error',
              metrics=['accuracy'])

learned = model.fit(X, y, epochs=10, batch_size=4, verbose=0)
print(model.predict(X))
```

## Plot Error

```
plt.plot(learned.history['loss'])
```
