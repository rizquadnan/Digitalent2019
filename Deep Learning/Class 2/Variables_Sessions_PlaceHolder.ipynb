# Variable and Session

```
import tensorflow as tf

const = tf.constant(2.0, name="const")

b = tf.Variable(2.0, name='b')
c = tf.Variable(1.0, name='c')

d = tf.add(b, c, name='d')
e = tf.add(c, const, name='e')
f = tf.multiply(d, e, name='f')

print(b)
```

```
with tf.Session() as sass:
  sass.run(tf.variables_initializer([b,c]))
  # sass.run(tf.global_variables_initializer())
  print(sass.run(b))
  print("b and c:", sass.run([b, c]))
  print("e (addition) f (multiplication): ", sass.run([e, f]))
```

# Placeholder

```
const = tf.constant(2.0, name="const")

b = tf.placeholder(tf.float32, name='b')
c = tf.placeholder(tf.float32, name='c')

d = tf.add(b, c, name='d')
e = tf.add(c, const, name='e')
f = tf.multiply(d, e, name='f')

print(b)
```

```
with tf.Session() as sass:
#   sass.run(tf.variables_initializer([b,c]))
  sass.run(tf.global_variables_initializer())
  print("d: ", sass.run(d, feed_dict={b: 2.0, c:1.0}))
```

# Automatic Function Differentiation

```
x = tf.constant(3.0)

with tf.GradientTape() as gd:
  gd.watch(x)
  y = x * x
  dy_dx = gd.gradient(y, x)
  
with tf.Session() as sass:
  out = sass.run(dy_dx)
  print(out)
```
