2. Write a Python program to read first 3 lines of a file

```python
with open(file_path, 'r') as file:
    test = file.readlines()
    test = test[:3]
print(test)
```

3. Write a Python program to append text to a file and display the text.

```python
with open(file_path, 'a') as file:
    file.write("\nBagus juga puisinya")

with open(file_path, 'r') as file:
    test = file.read()
print(test)
```
